## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# Assume sales amounts are uniformly distributed between $1000 and $5000
lower_bound, upper_bound = 1000, 5000
sample_sizes = [1, 2, 5, 10]  # Different sample sizes to demonstrate the CLT

for i, n in enumerate(sample_sizes):
    # Generating random daily sales data for 'n' days
    daily_sales = np.random.uniform(lower_bound, upper_bound, (10000, n))
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100

```python
def generate_uniform(bottom: int, ceiling: int, count:int):
    return np.random.uniform(bottom,ceiling,count)
```

```python
data1=generate_uniform(10,15,100)
plt.hist(data1, bins=10, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Uniform Distribution: 10-15, 100 samples')
plt.show()
```

```python
data2=generate_uniform(10,60,1000)
plt.hist(data2, bins=10, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Uniform Distribution: 10-50, 1000 samples')
plt.show()
```

How are the two distributions different?

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.

1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
def generate_normal(mean: float, std:float, count:int):
    return np.random.normal(loc=mean,scale=std,size=count)
```

```python
normal_data1=generate_normal(10,1,1000)
plt.figure(figsize=(8, 6))
plt.hist(normal_data1,bins=70,density=True, alpha=0.6, color='b')
plt.title('Normal Distribution for mean=10, std=1, count=1000')
plt.show()
```

```python
normal_data2=generate_normal(10,50,1000)
plt.figure(figsize=(8, 6))
plt.hist(normal_data2,bins=70,density=True, alpha=0.6, color='g')
plt.title('Normal Distribution for mean=10, std=50, count=1000')
plt.show()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
def generate_exponential(mean: float, size: int):
    scale = mean
    return np.random.exponential(scale=scale, size=size)

exp_data1 = generate_exponential(10, 10)
exp_data2 = generate_exponential(10, 100)  

plt.subplot(1, 2, 1)
plt.hist(exp_data1, bins=100, color="skyblue")
plt.title('Exponential Distribution (size=10)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(exp_data2, bins=100, color="g")
plt.title('Exponential Distribution (size=100)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

small samples are noisy and unreliable for seeing the distribution shape, while larger samples converge to the expected exponential curve.

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
mean=10
lam=1/mean
x=15

prob_analytical=1-math.exp(-lam*15)
print(f"P(X < 15) from formula = {prob_analytical:.4f} ({prob_analytical*100:.2f}%)")
```

```python
num_customers=1000
time_in_bank = np.random.exponential(scale=mean, size=num_customers)
prob_simulation = np.mean(time_in_bank < 15)
print(f"Mean time in bank = {np.mean(time_in_bank):.2f} minutes")
print(f"P(X < 15) from simulation = {prob_simulation:.4f} ({prob_simulation*100:.2f}%)")

x_vals = np.linspace(0, 60, 1000)
pdf_vals = lam * np.exp(-lam * x_vals)  

plt.hist(time_in_bank, bins=30, density=True, color='skyblue', label='Simulated data')
plt.title('Time Spent in Bank - Exponential Distribution (mean=10)')
plt.xlabel('Time (minutes)')
plt.ylabel('Density')
plt.show()
```

What is the probability that the customer will spend more than 15 minutes

```python
print(f"If P(X < 15) from formula = {prob_analytical:.4f} ({prob_analytical*100:.2f})%\nthen, P(X > 15) = {(1-prob_analytical):.4f}")
```

```python
prob_less = 1-math.exp(-lam*15)          
prob_more = math.exp(-lam*15)   
print(f"P(X < 15) = {prob_less:.4f} ({prob_less*100:.2f}%)")
print(f"P(X > 15) = {prob_more:.4f} ({prob_more*100:.2f}%)")

plt.hist(time_in_bank, bins=30, density=True, label='Simulated data')
plt.axvline(x=15, color='red', linestyle='--', label='x = 15')
plt.title('Time Spent in Bank - Exponential Distribution (mean=10)')
plt.xlabel('Time (minutes)')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()      
```
